```python
%%writefile /kaggle/working/ViT5_Base_QLoRa_VNDS/configs/baseline/training_configs.yaml.template
# Cấu hình Model & Dataset
model:
  model_name: "${MODEL_NAME:-VietAI/vit5-base}"
  dataset_name: "${DATASET_NAME:-nam194/vietnews}"

# Cấu hình lấy mẫu dữ liệu (Data Sampling)
data:
  train_samples: ${TRAIN_SAMPLES:-6000}
  val_samples: ${VAL_SAMPLES:-1000}

# Cấu hình kỹ thuật QLoRa
lora:
  lora_r: ${LORA_R:-8}
  lora_alpha: ${LORA_ALPHA:-16}
  lora_target_module: ${LORA_TARGET_MODULE:-['q', 'v']}
  lora_dropout: ${LORA_DROPOUT:-0.05}

# Cấu hình quá trình huấn luyện (Training Hyperparameters)
training:
  epochs: ${EPOCHS:-20}
  batch_size: ${BATCH_SIZE:-4}
  gradient_accumulation_steps: ${GRADIENT_ACCUMULATION_STEPS:-4}
  learning_rate: ${LEARNING_RATE:-0.0002}
  weight_decay: ${WEIGHT_DECAY:-0.01}
  warmup_ratio: ${WARMUP_RATIO:-0.05}
  label_smoothing_factor: ${LABEL_SMOOTHING_FACTOR:-0.1}
```

```python
%%writefile /kaggle/working/ViT5_Base_QLoRa_VNDS/runner/train_runner.sh
#!/bin/bash

# Dừng script ngay lập tức nếu có bất kỳ lỗi nào xảy ra
set -e 

echo "=== SETTING UP TRAINING EXPERIMENT ==="

# --- IMPORTANT PATHS ---
TEMPLATE_PATH="/kaggle/working/ViT5_Base_QLoRa_VNDS/configs/baseline/training_configs.yaml.template"
CONFIG_PATH="/kaggle/working/ViT5_Base_QLoRa_VNDS/configs/baseline/training_configs.yaml"
TRAIN_SCRIPT="/kaggle/working/ViT5_Base_QLoRa_VNDS/scripts/baseline/train.py"
USE_DORA=True                     
ADAPTER_DIR="OptimLR_QLODA_adapter"
CHECKPOINT_DIR="OptimLR_QLODA_checkpoint"
# --------------------------------

# --- HYPERPARAMS ADJUSTMENT ---
export LEARNING_RATE=0.0003143144372398535
# --------------------------------

# install gettext for bash
echo "Installing gettext library for bash"
apt-get install gettext

# install requirements.txt
echo "Installing requirements.txt"
pip install -qq -r requirements.txt

# Sinh file cấu hình hoàn chỉnh từ file yaml template
echo "Creating configs file at: $CONFIG_PATH"

(echo "cat <<EOF"; cat $TEMPLATE_PATH; echo ""; echo "EOF") | bash > $CONFIG_PATH

# Start training
echo "Start running $TRAIN_SCRIPT"

python $TRAIN_SCRIPT --config $CONFIG_PATH --use_dora $USE_DORA --adapter_dir $ADAPTER_DIR --checkpoint_dir $CHECKPOINT_DIR

echo "=== FINISH ==="
```
